# 0514-0516 Boston Hounsing

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

```python
# 각 data의 shape -> ex)13개로 이루어진 data가 404개 있다
print (x_train.shape)
print (y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# x data의 첫번째 요인(criminal rate)과 y_train을 plot
plt.plot(x_train[:,0], y_train, '.')
```

#### 사전 data에 대한 분석

 INPUT::
 1. crim: per capita crime rate by town.
 2. zn: proportion of residential land zoned for lots over 25,000 sq.ft.
 3. indus: proportion of non-retail business acres per town.
 4. chas: Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
 5. nox: nitrogen oxides concentration (parts per 10 million).
 6. rm: average number of rooms per dwelling.
 7. age: proportion of owner-occupied units built prior to 1940.
 8. dis: weighted mean of distances to five Boston employment centres.
 9. rad: index of accessibility to radial highways.
 10. tax: full-value property-tax rate per $10,000.
 11. ptratio: pupil-teacher ratio by town.
 12. black: 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
 13. lstat: lower status of the population (percent).

```python
boston_housing = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']

for i in range (13):
    plt.subplot(3, 5, i+1)
    plt.plot(x_train[:,i], y_train, '.')
    plt.title(boston_housing[i])
plt.show()
```

#### 기본적인 model

- 입력값 13, 출력값 1
- 훈련해야하는 parameter가 14인 이유: y = ax +1b 에서 b까지 포함

```python
model = Sequential()
model.add(Dense(1, input_shape=[13]))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

```python
model.fit(x_train, y_train, epochs=1000)
# epochs를 1000까지 늘리니 loss가 100000에서 26까지 떨어짐
#이를 더 떨어질수 있게 늘리는 방법은 hidden layer!
```

```python
model.evaluate(x_test, y_test)
```

#### hidden layer로 loss를 더 줄이기 위한 model

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
model.evaluate(x_test, y_test)
```

1. input shape가 (img_rows, img_cols), 즉 28x28의 matrix이기 때문에 이를 벡터값으로 만들기 위해서 flatten 사용

```python
img_rows = 28
img_cols = 28

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(img_rows, img_cols)),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0,2),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

1. 첫 번째 layer에서 output shape은 28x28의 784
2. 두 번째 layer에서 parameter은 (784+1)x512 = 401920
3. 마지막 layer에서는 (513+1)x10
