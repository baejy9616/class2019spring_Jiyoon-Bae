#### 네이버 기사 랭킹

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen("https://news.naver.com/main/ranking/popularDay.nhn?rankingType=popular_day&sectionId=100&date=20190325")
bsObject = BeautifulSoup(html, 'html.parser')
bsObject
```

```python
for n in range(0,30):
    li = bsObject.find('li', {'class':'ranking_item is_num'+str(n+1)})
    title = li.find('a').get('title')
    source = li.find('div',{"class":"ranking_office"}).get_text()
    view = li.find('div',{"class":"ranking_view"}).get_text()
    print(str(n+1)+'위', title,
          'from'+source, view+'뷰') 
```

엑셀 파일로 export

```python
with open('crawling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', '출처', 'view'])
        
    for n in range(0,30):
        li = bsObject.find('li', {'class':'ranking_item is_num'+str(n+1)})
        title = li.find('a').get('title')
        source = li.find('div',{"class":"ranking_office"}).get_text()
        view = li.find('div',{"class":"ranking_view"}).get_text()
        writer.writerow([n+1, title, source, view])
```
