```python
divs = bsObject.findAll("div",{"class":"ranks"})

for div in divs:
    ranks = div.findAll('i')
    for rank in ranks:
        number = rank.text
        title = div.next_sibling.next_sibling.find('b').get_text()
        print(number+'위', title)
      
```

```python
with open('crawling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title'])
        
    divs = bsObject.findAll("div",{"class":"ranks"})

    for div in divs:
        ranks = div.findAll('i')
        for rank in ranks:
            number = rank.text
            title = div.next_sibling.next_sibling.find('b').get_text()
            
            writer.writerow([number, title])
```

divs = bsObject.findAll("div",{"class":"ranks"})
tds = bsObject.find_all('td', {"class":"prdDuration"})

for div in divs:
    ranks = div.findAll('i')
    for rank in ranks:
        number = rank.text
        title = div.next_sibling.next_sibling.find('b').get_text()
    for td in tds:
        b = td.next_sibling.next_sibling.findAll('b')
        for per in b:
            perc = per.get_text()
                
            print(number+'위', title, perc)
